```python
# Handle Missing Values
# Separate numerical and categorical columns
numerical_cols = train_df.select_dtypes(include=['float64', 'int64']).columns.drop('SalePrice')  # Exclude target column
categorical_cols = train_df.select_dtypes(include=['object']).columns

# Impute missing values
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

# Apply imputers to train and test datasets
train_df[numerical_cols] = num_imputer.fit_transform(train_df[numerical_cols])
train_df[categorical_cols] = cat_imputer.fit_transform(train_df[categorical_cols])

test_df[numerical_cols] = num_imputer.transform(test_df[numerical_cols])
test_df[categorical_cols] = cat_imputer.transform(test_df[categorical_cols])
```

Code ini digunakan untuk menangani nilai yang hilang (missing values) dalam dataset, baik pada data pelatihan (train_df) maupun data uji (test_df). Proses imputasi dilakukan dengan cara mengganti nilai yang hilang pada kolom numerik menggunakan nilai median dan pada kolom kategorikal menggunakan nilai yang paling sering muncul. Dengan demikian, dataset menjadi lengkap tanpa nilai hilang dan siap untuk digunakan dalam pelatihan model.

```python
def feat_eng(data):
    """
    Feature Engineering untuk dataset rumah.
    """
    # Fitur Luas
    data['TotalSF'] = data['TotalBsmtSF'] + data['1stFlrSF'] + data['2ndFlrSF']
    data['TotalArea'] = data['GrLivArea'] + data['TotalBsmtSF']
    data['TotalPorchSF'] = data['OpenPorchSF'] + data['EnclosedPorch'] + \
                           data['3SsnPorch'] + data['ScreenPorch'] + data['WoodDeckSF']
    data['TotalOutdoorSF'] = data['WoodDeckSF'] + data['TotalPorchSF']
    data['FinishedSF'] = data['BsmtFinSF1'] + data['BsmtFinSF2'] + data['GrLivArea']

    # Fitur Kamar Mandi
    data['TotalBath'] = data['FullBath'] + (0.5 * data['HalfBath']) + \
                        data['BsmtFullBath'] + (0.5 * data['BsmtHalfBath'])

    # Fitur Umur dan Renovasi
    data['HouseAge'] = data['YrSold'] - data['YearBuilt']
    data['SinceRenovation'] = data['YrSold'] - data['YearRemodAdd']
    data['GarageAge'] = data['YrSold'] - data['GarageYrBlt']
    data['GarageAge'] = data['GarageAge'].fillna(data['GarageAge'].max())  # Isi NaN dengan umur maksimum
    data['IsRemodeled'] = (data['YearBuilt'] != data['YearRemodAdd']).astype(int)

    # Fitur Kualitas
    quality_mapping = {'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1}
    data['OverallScore'] = data['OverallQual'] * data['OverallCond']
    data['ExterScore'] = data['ExterQual'].map(quality_mapping) + data['ExterCond'].map(quality_mapping)
    data['BsmtScore'] = data['BsmtQual'].map(quality_mapping).fillna(0) + \
                        data['BsmtCond'].map(quality_mapping).fillna(0)
    data['GarageScore'] = data['GarageQual'].map(quality_mapping).fillna(0) + \
                          data['GarageCond'].map(quality_mapping).fillna(0)

    # Fitur Kamar dan Ruangan
    data['TotalRooms'] = data['TotRmsAbvGrd'] + data['FullBath'] + data['HalfBath']
    data['IsBigHouse'] = (data['GrLivArea'] > 2500).astype(int)

    # Fitur Rasio
    data['GrLivAreaRatio'] = data['GrLivArea'] / data['LotArea']
    data['BasementRatio'] = data['TotalBsmtSF'] / data['TotalArea']
    data['GarageToTotalArea'] = data['GarageArea'] / data['TotalArea']
    data['GarageToTotalArea'] = data['GarageToTotalArea'].fillna(0)

    # Fitur Indikator
    data['HasBasement'] = (data['TotalBsmtSF'] > 0).astype(int)
    data['HasPool'] = (data['PoolArea'] > 0).astype(int)
    data['Has2ndFloor'] = (data['2ndFlrSF'] > 0).astype(int)
    data['IsFlatLand'] = (data['LandSlope'] == 'Gtl').astype(int)
    data['HasFullUtilities'] = (data['Utilities'] == 'AllPub').astype(int)
    data['IsPremiumLot'] = ((data['LotConfig'] == 'Corner') | (data['LotConfig'] == 'CulDSac')).astype(int)
    data['IsNewHouse'] = (data['YrSold'] - data['YearBuilt'] <= 5).astype(int)  # Rumah baru

    # Fitur Musim
    data['SeasonSold'] = data['MoSold'].map({
        12: 'Winter', 1: 'Winter', 2: 'Winter',
        3: 'Spring', 4: 'Spring', 5: 'Spring',
        6: 'Summer', 7: 'Summer', 8: 'Summer',
        9: 'Fall', 10: 'Fall', 11: 'Fall'
    })
    one_hot = pd.get_dummies(data['SeasonSold'], prefix='SeasonSold')
    data = pd.concat([data, one_hot], axis=1)
    data.drop('SeasonSold', axis=1, inplace=True)

    # Fitur Baru Berdasarkan Zonasi
    data['IsResidential'] = (data['MSZoning'] == 'RL').astype(int)  # Residential Large Lot
    data['HighValueNeighborhood'] = data['Neighborhood'].isin(['StoneBr', 'NoRidge', 'NridgHt']).astype(int)

    # Fitur Ekstra
    data['YearsSinceGarageBuilt'] = data['YrSold'] - data['GarageYrBlt']
    data['YearsSinceGarageBuilt'] = data['YearsSinceGarageBuilt'].fillna(data['YearsSinceGarageBuilt'].max())

    # Cek nilai NaN setelah mapping
    data['ExterScore'] = data['ExterScore'].fillna(0)
    data['GarageScore'] = data['GarageScore'].fillna(0)

    return data

# Terapkan feature engineering pada data train dan test
train_df = feat_eng(train_df)
test_df = feat_eng(test_df)
```

Code ini melakukan Feature Engineering pada dataset rumah untuk menciptakan fitur-fitur baru yang dapat meningkatkan kinerja model prediksi. Berikut adalah fitur-fitur yang dibuat:

- Fitur Luas: Menghitung total luas berbagai bagian rumah, seperti luas basement, lantai pertama, lantai kedua, dan total luas outdoor.
- Fitur Kamar Mandi: Menghitung total kamar mandi berdasarkan jumlah full bath, half bath, dan basement bath.
- Fitur Umur: Menambahkan fitur umur rumah, umur renovasi, dan umur garasi.
Fitur Kualitas: Mengkombinasikan nilai kualitas eksternal dan basement untuk memberikan skor kualitas keseluruhan rumah.
- Fitur Kamar dan Ruangan: Menghitung total ruangan dan menambahkan indikator rumah besar.
- Fitur Rasio: Menambahkan rasio ruang hidup (GrLivArea) terhadap area tanah (LotArea), basement terhadap total area, dan garasi terhadap total area.
- Fitur Indikator: Menambahkan indikator apakah rumah memiliki basement, kolam renang, lantai kedua, dan lain-lain.
- Fitur Musim: Menambahkan fitur yang menggambarkan musim saat rumah dijual berdasarkan bulan.

Setelah itu, data yang telah diubah akan kembali diterapkan pada dataset pelatihan (train_df) dan dataset pengujian (test_df) untuk digunakan dalam pelatihan model.

Tujuan dari feature engineering ini adalah untuk membuat dataset lebih kaya akan informasi dan memberikan model prediksi lebih banyak fitur yang relevan.

# **Cross Validatuon**

```python
# Cross Validation
import numpy as np
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error

# Inisialisasi model dengan parameter terbaik
params = {
    'iterations': 920,
    'depth': 5,
    'learning_rate': 0.1644637010279701,
    'l2_leaf_reg': 5.745394442153676,
    'random_strength': 1.9494141346574156,
    'bagging_temperature': 0.6004041752712852
}
model = CatBoostRegressor(random_state=42, verbose=0)

# Definisi fungsi scoring RMSE untuk cross-validation
def rmse_scorer(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Menggunakan make_scorer untuk konversi ke fungsi sklearn-compatible
rmse_score = make_scorer(rmse_scorer, greater_is_better=False)

# Cross-validation dengan 5 fold (dapat disesuaikan sesuai kebutuhan)
cv_scores = cross_val_score(model, X, y, cv=5, scoring=rmse_score)

# Output hasil RMSE per fold dan rata-rata
print("Cross-validation RMSE per fold:", -cv_scores)
print("Average RMSE:", -np.mean(cv_scores))
```


Hasil cross-validation untuk model CatBoost menunjukkan RMSE per fold yang bervariasi antara 20,220.26 hingga 30,319.49, dengan rata-rata RMSE sebesar 25,472.09. Ini menunjukkan bahwa model ini cukup konsisten dalam memprediksi, dengan performa terbaik pada beberapa fold dan sedikit lebih buruk pada yang lainnya.
